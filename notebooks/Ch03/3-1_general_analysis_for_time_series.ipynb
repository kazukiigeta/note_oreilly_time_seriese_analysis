# 3.1 よく使用される手法

## データの確認
1991-1998年における4つの欧州主要株価指数の営業日ごとの終値のデータセットである[EuStockMarkets](https://www.rdocumentation.org/packages/datasets/versions/3.6.2/topics/EuStockMarkets)を使う。

```r
head(EuStockMarkets)
```

## プロット
時系列データは順番に意味があるため単にプロットを行うだけで多くの情報が得られる。

```r
plot(EuStockMarkets)
```

`plot()`だけでプロットが4つの時系列に分かれるのは、EuStockMarketsがmtsクラスのオブジェクトだからである。

```r
class(EuStockMarkets)
```

`ts`オブジェクトには以下の便利な関数があるが、インデックス付けがやりにくい場合もある。

- `frequency()`で1年あたりの頻度を求める。

```r
frequency(EuStockMarkets)
```

- `start()`と`end()`で時系列の最初と最後の時刻を求める。

```r
start(EuStockMarkets)
end(EuStockMarkets)
```

- `window()`でデータの時間断面を抽出する。

```r
window(EuStockMarkets, start = 1997, end = 1998)
```

## ヒストグラム

元データにトレンドがあることから予測できる通り、変換前のヒストグラムは幅が広く正規分布を示さない。

ラグ1の階差を取ることでトレンドを除去すると、データはより正規分布に近づく。

階差系列では、値そのものではなく値の上下が示されており、実際の値よりも情報量が多い。

階差系列のヒストグラムでは正の値に少しだけバイアスが存在していることで上向きのトレンドになっていることがわかる。

```r
hist(     EuStockMarkets[, "SMI"], 30)
hist(diff(EuStockMarkets[, "SMI"], 30))
plot(diff(EuStockMarkets[, "SMI"], 30))
```

## 散布図

散布図は他の種類のデータと同じくらい時系列にも有用。

以下の2つの事例では相関が非常に強いように見えるが、関係性は見えるほど強くない（疑似相関）。

- 長期にわたる2つの株価
- 長期にわたる2つの株式の日々の変動

2つのプロットの相関は、同じ時間点での相関であるために株式トレーダーが収益化できるような情報とはならない（SMIの値からDAXを予測するようなことが出来ない）。

```r
plot(EuStockMarkets[, "SMI"], EuStockMarkets[, "DAX"])
plot(diff(EuStockMarkets[, "SMI"]), diff(EuStockMarkets[, "DAX"]))
```

SMIの階差値を1つ前にシフトして散布図を描くと、株式間の相関は消滅する。

このことから、SMIの株価からDAXの株価を予測できないだろうことが分かる。

```r
plot(lag(diff(EuStockMarkets[, "SMI"]), 1), diff(EuStockMarkets[, "DAX"]))
```

以上より、以下の2点が分かる。

- 時系列データには時系列でないデータと同じ探索的データ解析手法が使えるが、データの形を変換してから適用する必要がある。
- 多くの場合、データの挙動に関する情報量が最も多いのは、異なる時間点でのデータの関係や、データの時間変化である。
